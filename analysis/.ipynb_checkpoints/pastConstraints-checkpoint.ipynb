```python
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

sb.set(style='white', context='talk', palette='Dark2')
plt.rcParams["font.family"] = "serif"
plt.rcParams["figure.dpi"] = 300
plt.rcParams["axes.grid.which"] = 'both'
plt.rcParams['ytick.left'] = True
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.right'] = True
plt.rcParams['xtick.top'] = True
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.direction'] = 'in'
```

```python
plotConstraint=True

fig, ax = plt.subplots(figsize=(8,6))

mus = [1.2, 2, 1, 4.5, 4, 2.2, 2.6, 0.5, 31]
# Capozzi & Raffelt (2020)
# Raffelt 1990
# Castellani & DeglInnocenti (1993)
# Viaux et al (2013)
# Straniero et al (2018)
# Arceo-Daz et al (2015)
# Arceo-Daz et al (2019)
# WDLF : Bertolami & Miguel (2014)
labels = ['C&R20', 'R90', 'C&D93', 'V13', 'S18', 'A19', 'A15', 'WDLF', 'Solar']
shapes = ['o', 'P', '^', 'v', '*', 's', 'X', '>', '<']

x = np.linspace(0, len(mus)+1, len(mus))

for ii in range(len(mus)):
    ax.plot([x[ii]], [mus[ii]], shapes[ii], markersize=20)

ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=-60)
ax.set_ylabel(r'$\mu_{12}$')
ax.tick_params(which='major', length=7)
ax.set_yscale('log')
ax.axvline(np.mean([x[-2], x[-3]]), ymin=-0.3, ymax=1, clip_on=False, color='k', linestyle=':')
ax.text(2.0, 10**-0.9, 'TRGB Constraints')
ax.text(np.mean([x[-1], x[-2]])-1, 10**-1, 'Other\nConstraints')

if plotConstraint:
    ax.axhline(6, color='k', alpha=0.5, linestyle='--')
    ax.fill_between(np.linspace(-1, len(mus)+2, 100), 0*np.ones(100), 6*np.ones(100), facecolor="none", hatch="X", edgecolor="b", alpha=0.5, linewidth=0.0)
    ax.set_xlim(-0.5, len(mus)+1.5)
    ax.set_ylim(0, max(mus)+10)
```
