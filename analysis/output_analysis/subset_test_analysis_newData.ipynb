```python
import os, sys, glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sb

sb.set(context='talk', style='whitegrid', palette='Set1')

def plotMI(mesaOutFile):
    mags = pd.read_csv(mesaOutFile, index_col=0)
    print(mags.keys())
    print(mags[['flag', 'mass_index', 'y_index', 'z_index', 'mu_index', 'M_I']])
    
    magsGood = mags[mags['flag'] != 1]
    
    fig, ax = plt.subplots()
    ax.plot(magsGood.mu, magsGood.M_I, 'o')
    group = mags.groupby(mags.mu).mean().reset_index()
    #ax.plot(group.mu,group.M_I, 'o')
    ax.set_xlabel('Mu_12')
    ax.set_ylabel('I-Band Magnitude')
    
    # Histogram the flags
    fig, ax = plt.subplots()
    ax.hist(mags.flag)
    ax.set_xlabel('Flag')
    
```
