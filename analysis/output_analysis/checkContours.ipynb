```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sb
sb.set(context='talk', palette='Set1', style='white')
```

```python
gridfile = 'allData.csv'
grid = pd.read_csv(gridfile, index_col=0)
grid.z = np.log10(grid.z)

obsI = -4.047
obsErr = 0.045

grid = grid[(grid.M_I > (obsI-obsErr)) & (grid.M_I < (obsI+obsErr))]
```

```python
mu2 = np.sort(grid.mu.unique())[-2]
mu1 = np.sort(grid.mu.unique())[-6]

grid1 = grid[grid.mu == mu1]
grid2 = grid[grid.mu == mu2]
```

```python
fig, ax = plt.subplots(figsize=(8,6))

#M1, Z1 = np.meshgrid(grid1.mass, grid1.z)
ms = 12
ax.plot(grid1.mass, grid1.z, 'o', alpha=0.5, markersize=ms, label=r'$\mu_{12}=$' + )
ax.plot(grid2.mass, grid2.z, 'X', alpha=0.5, markersize=ms, label=r'$\mu_{12}$ consistent w/ Z')
ax.legend()
ax.set_ylabel(r'log$_{10}$(Z)')
ax.set_xlabel('M');
```
