```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sb
sb.set(context='talk', palette='Set1', style='white')
```

```python
gridfile = 'allData.csv'
grid = pd.read_csv(gridfile, index_col=0)
grid = grid[grid.mu != 0]
```

```python
obsI = -4.047
obsErr = 0.045

grid = grid[(grid.M_I > (obsI-obsErr)) & (grid.M_I < (obsI+obsErr))]

print(np.sort(grid.mass.unique()))
mass1 = np.sort(grid.mass.unique())[-5]
mass2 = np.sort(grid.mass.unique())[0]

grid1 = grid[grid.mass == mass1]
grid2 = grid[grid.mass == mass2]
```

```python
fig, ax = plt.subplots(figsize=(8,6))

#M1, Z1 = np.meshgrid(grid1.mass, grid1.z)
ms = 12
ax.plot(grid2.z, grid2.mu, 'X', alpha=0.5, markersize=ms, label=r'mass=' + str(round(mass2, 4)))
ax.plot(grid1.z, grid1.mu, 'o', alpha=0.5, markersize=ms, label=r'mass=' + str(round(mass1, 4)))

ax.legend()
ax.set_xlabel('Z')
ax.set_ylabel(r'$\mu_{12}$');
```
