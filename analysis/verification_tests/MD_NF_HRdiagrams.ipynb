```python
import os, glob
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

from mesa_reader import MesaData

sb.set(style='whitegrid', context='talk', palette='Paired')
```

```python
datadir = '/home/ubuntu/Documents/NMDinStars/mesa_mod/mesa-r12778/star/test_suite/NuetrinoMagneticDipoleMesa/LOGS/'

myData = glob.glob(datadir+'*.data')
mdData = glob.glob(datadir+'MDtests/*.data')

fig, ax = plt.subplots(1, figsize=(8,6))

for nf, md in zip(myData, mdData):
    
    m_nf = MesaData(nf)
    m_md = MesaData(md)
    ax.plot(m_nf.effective_T, m_nf.log_L, alpha=0.5)
    ax.plot(m_md.effective_T, m_md.log_L, alpha=0.5)

ax.set_xlabel('Effective Temperature [K]')
ax.set_ylabel('log(L)')
plt.gca().invert_xaxis()
```
